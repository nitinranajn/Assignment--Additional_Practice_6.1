```python
#### Get Data from the following link: http://files.grouplens.org/datasets/movielens/ml-20m.zip
#### We will be using the following files for this exercise:
# ratings.csv : userId,movieId,rating, timestamp
# tags.csv : userId,movieId, tag, timestamp
# movies.csv : movieId, title, genres
import os
import zipfile as zp

os.system("wget 'http://files.grouplens.org/datasets/movielens/ml-20m.zip'") 
zip= zp.ZipFile('ml-20m.zip')
zip.extractall() #unzip the folder
os.listdir('C:\\Users\\nitin\\ml-20m') # List the file name from ml-20m folder
```

```python
#1. Read the dataset using pandas.
import pandas as pd
movies = pd.read_csv('./ml-20m/movies.csv', sep=',')
tags = pd.read_csv('./ml-20m/tags.csv', sep=',')
ratings = pd.read_csv('./ml-20m/ratings.csv', sep=',')
```

```python
# 5. Calculate descriptive statistics for the 'ratings' column of the ratings DataFrame. Verify using describe().

print("Count = ", ratings['rating'].count() )
print("Mean = ", ratings['rating'].mean())
print("Std Daviaton  = ", ratings['rating'].std())
print("max value = ", ratings['rating'].max())
print("Minimum value = ", ratings['rating'].min())

print("\nFind the statistics of rating datframe using describe()\n+++++++++++++++++++++++++++++++++++++++++++\n")
print("statistics for the 'ratings' column:  \n", ratings['rating'].describe())
# print("\nStatistics for rating dataframe:  \n", ratings.describe())
```

```python
# 6. Filter out ratings with rating > 5
values = ratings['rating']>5
print(values.any())
```

```python
#removing missing values
tags=tags.dropna()
tags.isnull().any()
```

```python
tags.shape
```

```python
# 8. Filter out movies from the movies DataFrame that are of type 'Animation'.
movie= movies['genres'].str.contains('Animation')
movies[movie][1:6]
```

```python
# 9. Find the average rating of movies.
av_rating = ratings[['movieId','rating']].groupby('movieId').mean()
av_rating.head()
```

```python
# 10. Perform an inner join of movies and tags based on movieId.
join= movies.merge(tags, on='movieId', how='inner')
join.head()
```

```python
# 11. Print out the 5 movies that belong to the Comedy genre and have rating greater than 4.
avg_ratings = ratings.groupby('movieId', as_index=False).mean()
box_office = movies.merge(avg_ratings, on='movieId', how='inner')
is_comedy= box_office['genres'].str.contains('Comedy')
is_highly_rated = box_office['rating'] >= 4.0
box_office[is_comedy & is_highly_rated].head()
```

```python
# 12. Split 'genres' into multiple columns.
movie_genres = movies['genres'].str.split('|', expand=True)
movie_genres.head()
```

```python
#13. Extract year from title e.g. (1995).
movies['year'] = movies['title'].str.extract('.*\((.*)\).*', expand=True)
movies.head()
```

```python
# 14. Select rows based on timestamps later than 2015-02-01.
tags['parsed_time'] = pd.to_datetime(tags['timestamp'], unit='s')
greater_than_t = tags['parsed_time'] > '2015-02-01'
selected_rows = tags[greater_than_t]
print(selected_rows.head())
tags.shape, selected_rows.shape
```

```python
# 15. Sort the tags DataFrame based on timestamp.
tags.sort_values(by='timestamp', ascending=True).head(20)
```
